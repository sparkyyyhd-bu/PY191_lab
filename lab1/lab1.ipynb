```python
# Your code here
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

print(factorial(5))
```

```python
import timeit
import matplotlib.pyplot as plt

def traditional_for():
    l = []
    for i in range(10000):
        l.append(i**2)
    return l

def list_comprehension():
    l = [i**2 for i in range(10000)]
    return l

time1 = timeit.timeit(traditional_for, number = 1)
time2 = timeit.timeit(list_comprehension, number = 1)

plt.bar(["Traditional For-Loop", "List Comprehension"], [time1, time2])
plt.ylabel("Time (s)")

```
