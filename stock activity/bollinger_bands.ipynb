```python
import pandas as pd
unh = pd.read_csv("/Users/lucaslu/Lucas/Code Repositories/PY191_lab/unh.csv")
unh.drop(unh.index[0], inplace=True)
print(unh)
```

```python
money = 100.0
shares = 0.0
sma_higher = False
trades = 0
for row in unh.itertuples(index=True):
    prev = sma_higher
    sma_higher = row.Short_MA > row.Long_MA
    if not (sma_higher == prev):
        trades += 1
        if sma_higher:
            shares += money / float(row.Close)
            money = 0
        else:
            money += shares * float(row.Close)
            shares = 0
if money == 0:
    money += shares * float(row.Close)
    shares = 0

print(f"Number of trades: {trades}")
print(f"Simplified Bollinger Bands: {money}")
print(f"Buy and hold: {100*float(unh.iloc[-1].Close) / float(unh.iloc[0].Close)}")
```
