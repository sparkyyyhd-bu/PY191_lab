```python
import pandas as pd
unh = pd.read_csv("/Users/lucaslu/Lucas/Code Repositories/PY191_lab/unh.csv")
unh.drop(unh.index[0], inplace=True)
print(unh)

spy = pd.read_csv("/Users/lucaslu/Lucas/Code Repositories/PY191_lab/spy.csv")
spy.drop(spy.index[0], inplace=True)
print(spy)
```

```python
unh_return = unh["Return"]
spy_return = spy["Return"]
print(unh_return, spy_return)
```

```python
import numpy as np
beta, alpha = np.polyfit(spy_return, unh_return, 1)
print(f"Slope (m): {beta:.4f}")
print(f"Intercept (c): {alpha:.4f}")

matrix = np.corrcoef(spy_return, unh_return)
r = matrix[0, 1]

print(f"Correlation coefficient (r): {r}")

```
